```python
import pandas as pd

# Create a dataframe
data = pd.DataFrame({'First name and last name': ['John Doe', 'Jane Doe', 'Peter Smith'], 'GPA': [3.5, 3.8, 2.9]})

# Count unique students
unique_students = len(data['First name and last name'].unique())
print(f"Number of unique students: {unique_students}")

# Calculate average GPA
average_gpa = data['GPA'].mean()
print(f"Average GPA: {average_gpa:.2f}")

```

DISTRIBUTION OF GRADUATION YEARS AND TOP 5 COLLEGES BY GPA

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe
data = pd.DataFrame({'Year of graduation': [2019, 2020, 2021, 2022, 2023], 'College name': ['Stanford', 'Harvard', 'MIT', 'Caltech', 'UC Berkeley'], 'GPA': [3.9, 3.8, 3.7, 3.6, 3.5]})

# Distribution of graduation years
plt.figure(figsize=(8, 6))
data['Year of graduation'].hist(bins=10, edgecolor='black')
plt.title('Distribution of Graduation Years')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()

# Top 5 colleges by GPA
top_colleges = data.groupby('College name')['GPA'].mean().nlargest(5)
plt.figure(figsize=(10, 6))
top_colleges.plot(kind='bar', color='skyblue')
plt.title('Top 5 Colleges by Average GPA')
plt.xlabel('College')
plt.ylabel('Average GPA')
plt.show()

```

SPOTTING RELATIONSHIPS

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe
data = pd.DataFrame({'Family income': [100000, 120000, 150000, 200000, 250000],
                    'GPA': [3.5, 3.8, 3.7, 3.6, 3.4],
                    'Leadership skills': [5, 4, 3, 2, 1],
                    'Expected salary': [100000, 120000, 150000, 200000, 250000]})

# Relationship between family income and GPA
plt.figure(figsize=(8, 6))
plt.scatter(data['Family income'], data['GPA'], alpha=0.5)
plt.title('Family Income vs. GPA')
plt.xlabel('Family Income')
plt.ylabel('GPA')
plt.show()


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe
data = pd.DataFrame({'Leadership skills': [5, 4, 3, 2, 1],
                    'Expected salary': [100000, 120000, 150000, 200000, 250000]})

# Relationship between leadership skills and expected salary
plt.figure(figsize=(8, 6))
plt.scatter(data['Leadership skills'], data['Expected salary'], alpha=0.5)
plt.title('Leadership Skills vs. Expected Salary')
plt.xlabel('Leadership Skills')
plt.ylabel('Expected Salary')

# Add a linear regression line
slope, intercept = np.polyfit(data['Leadership skills'], data['Expected salary'], 1)
plt.plot(data['Leadership skills'], slope * data['Leadership skills'] + intercept, color='red')

plt.show()

```

DATA VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataframe
data = pd.DataFrame({'City': ['Bangalore', 'Mumbai', 'Delhi', 'Chennai', 'Kolkata'],
                    'GPA': [8, 7, 9, 6, 8],
                    'Expected salary': [100000, 120000, 150000, 200000, 250000],
                    'Experience with Python programming (in months)': [12, 24, 36, 18, 24]})

# Student distribution across cities
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='City', palette='Set3')
plt.title('Student Distribution Across Cities')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.show()


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataframe
data = pd.DataFrame({'GPA': [8, 7, 9, 6, 8],
                    'Expected salary': [100000, 120000, 150000, 200000, 250000],
                    'Experience with Python programming (in months)': [12, 24, 36, 18, 24]})

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataframe
data = pd.DataFrame({'GPA': [8, 7, 9, 6, 8],
                    'Experience with Python programming (in months)': [12, 24, 36, 18, 24]})
data['Expected Salary'] = [100000, 120000, 150000, 200000, 250000]

# Expected salary variations based on GPA, family income, and Python experience
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GPA', y='Expected Salary', hue='Experience with Python programming (in months)')
plt.title('Expected Salary Variation')
plt.xlabel('GPA')
plt.ylabel('Expected Salary')
plt.legend(title='Python Experience')
plt.show()
```
